```python
import pandas as Fiddu
import matplotlib.pyplot as adc
lr=Fiddu.read_csv('/content/energy.csv')
print(lr)
srh=lr[lr['State']=='Andhra']
rcb=lr[lr['State']=='Karnataka']
fig,diag=adc.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'])
diag[1].plot(rcb['Date'],rcb['Consumption'])
#bar chart
diag[0].bar(srh['Date'],srh['Consumption'])
diag[1].bar(rcb['Date'],rcb['Consumption'])

#pie chart
csds='Kerala'
cse='Delhi'
ece=[csds,cse]
tdp=lr[lr['State']=='Kerala']['Consumption'].sum()
ysr=lr[lr['State']=='Delhi']['Consumption'].sum()
jsp=[tdp,ysr]
fig,diag=adc.subplots(2)
diag[0].pie(jsp,labels=ece,autopct='%1.1f%%')
```
